```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

def generate_research_book_name(query, model_name="gpt2"):
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)
    model = GPT2LMHeadModel.from_pretrained(model_name)

    # Combine query and book name structure in the prompt
    prompt = f"Query: \"{query}\" Research Book Name: "

    # Tokenize and generate output
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(input_ids, max_length=100, num_beams=5, temperature=0.7, top_k=50, top_p=0.95, length_penalty=2.0, num_return_sequences=1)

    # Decode the generated output
    book_name = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return book_name

query = "Is a last will considered the final will of a person?"
research_book_name = generate_research_book_name(query, model_name="gpt2")
print("Query:", query)
print("Generated Research Book Name:", research_book_name)
```

```python
from transformers import pipeline

def generate_research_book_name(query, model_name="MBZUAI/LaMini-Flan-T5-248M"):
    text2text_generator = pipeline("text2text-generation", model=model_name)

    # Combine query and book name structure in the prompt
    prompt = f"Query: \"{query}\" Research Book Name:"

    # Generate output
    generated_output = text2text_generator(prompt, max_length=100, num_return_sequences=1)

    # Extract the generated book name from the output
    book_name = generated_output[0]['generated_text'].replace(prompt, "").strip()

    return book_name

query = "Is a last will considered the final will of a person?"
research_book_name = generate_research_book_name(query, model_name="MBZUAI/LaMini-Flan-T5-248M")
print("Query:", query)
print("Generated Research Book Name:", research_book_name)
```
